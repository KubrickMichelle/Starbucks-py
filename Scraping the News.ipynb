```python
url = 'http://uk.reuters.com/business/quotes/companyNews?symbol=SBUX.OQ'
```

```python
from urllib.parse import urlparse
from bs4 import BeautifulSoup
import pyodbc
```

```python
from urllib import request
```

```python
req = request.urlopen(url)
```

```python
req.read
```

```python
content = req.read()
```

```python
soup = BeautifulSoup(content, 'html.parser')
```

```python
soup.title
```

```python
soup
```

```python
headlines = soup.find_all('h2')
       
```

```python
headlines
```

```python
finalheadline = []
for h in headlines:
    hs = str(h)
    hl = hs.split('>')[2].replace('</a','').replace('2-','').replace('1-','')
    finalheadline.append(hl)
   
```

```python
finalheadline
```

```python
description = soup.find_all('p')
```

```python
description
```

```python
finaldescription = []
for d in description:
    ds = str(d)
    dl = ds.split('>')[1].replace('</p','').replace('July 27 ','').replace('July 6 ','').replace('* ','').replace('\'','')
    finaldescription.append(dl)
```

```python
finaldescription
```

```python
conn = pyodbc.connect(r'DSN=costa')
curr= conn.cursor()
```

```python
news = {}
```

```python
for x in range(0,len(finalheadline)):
    news[finalheadline[x]] = finaldescription[x]
```

```python
news
```

```python
for k,v in news.items():
    q = 'insert into News2(\
  Headline, [Description], [Time] )\
  values (\'%s\',\'%s\', getdate())' \
  %( k,v)
    try:
        curr.execute(q)
        conn.commit()
    except:
        print(q) 
```
